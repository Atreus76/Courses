```python
new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
```

#### TASK: Lật ngược ảnh và hiển thị trong notebook.

```python
img_v = cv2.flip(new_img,0)
plt.imshow(img_v)
```

#### TASK: Vẽ một hình chữ nhật trống MÀU ĐỎ quanh khuôn mặt của chú chó và hiển thị trong notebook.

```python
rec_img = cv2.rectangle(new_img, pt1=(200,380), pt2=(600,700), color=(255,0,0), thickness=10)
plt.imshow(rec_img)
```

#### TASK: Vẽ một HÌNH TAM GIÁC MÀU XANH DƯƠNG ở giữa ảnh. Có thể chọn kích thước và góc tùy ý, nhưng phải là hình tam giác (3 cạnh) theo bất kỳ hướng nào.

```python
vertices = np.array([[250,700],[400,400],[550,700]])
pts = vertices.reshape((-1,1,2))
triangle = cv2.polylines(new_img, [pts], isClosed=True, color=(0,0,255), thickness=10)
plt.imshow(triangle)
```

### BONUS TASK. Bạn có thể tô màu tam giác này không? Điều này cần một hàm khác, không có trong bài giảng. Hãy thử xem liệu bạn có thể tìm thấy trên Google không.

[CLICK VÀO LIÊN KẾT SAU ĐỂ XEM GỢI Ý](https://docs.opencv.org/3.0-beta/modules/imgproc/doc/drawing_functions.html#fillpoly)

```python
fill_triangle = cv2.fillPoly(triangle, [pts], color=(0,0,255))
plt.imshow(fill_triangle)
```

#### TASK: (LƯU Ý: BẠN SẼ CẦN CHẠY ĐIỀU NÀY LÀM SCRIPT).Tạo một script để mở ảnh và cho phép vẽ hình tròn trống màu đỏ mỗi khi bạn click CHUỘT PHẢI.

```python
def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(image,(x,y),100,(0,0,255),-1)

cv2.namedWindow(winname='my_drawing')
cv2.setMouseCallback('my_drawing', draw_circle)

image = np.zeros((512,512,3), np.int8)
while True:
    cv2_imshow('my_drawing', image)
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
