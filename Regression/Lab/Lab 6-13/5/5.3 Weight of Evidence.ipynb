## Trọng số bằng chứng

Trọng số bằng chứng (WoE) được phát triển chủ yếu cho ngành tài chính và tín dụng, giúp xây dựng nhiều mô hình có tính dự báo hơn để đánh giá rủi ro vỡ nợ. Nó dự đoán khả năng số tiền cho một người hoặc tổ chức vay bị mất. Như vậy, trọng số bằng chứng là thước đo “sức mạnh” của kỹ thuật phân nhóm để phân tách rủi ro tốt và xấu (vỡ nợ).

- WoE sẽ bằng 0 nếu P (Goods)/P (Bads) = 1, nghĩa là, nếu kết quả là ngẫu nhiên cho nhóm đó.
- Nếu P(Bads) > P(Goods) thì tỷ lệ chênh lệch sẽ < 1 và
- WoE sẽ < 0 nếu P (Goods)> P (Bads).

WoE rất phù hợp với Hồi quy Logistic vì phép biến đổi Logit chỉ đơn giản là log của tỷ lệ chêch lệch, tức là ln(P(Goods)/P(Bads)). Do đó, bằng cách sử dụng các yếu tố dự đoán WoE được mã hóa trong hồi quy logistic, tất cả các yếu tố dự báo đã được chuẩn bị và mã hóa theo cùng một tỷ lệ, và các tham số trong phương trình hồi quy logistic tuyến tính có thể so sánh trực tiếp.

Phép biến đổi WoE có 3 ưu điểm:

- Nó tạo ra một mối quan hệ đơn điệu giữa mục tiêu và các biến độc lập.
- Nó sắp xếp các hạng mục theo thang đo "logistic", điều này là tự nhiên đối với hồi quy logistic.
- Sau đó các biến có thể được so sánh vì chúng có cùng thang đo nên có thể xác định biến nào mang tính dự đoán cao hơn.

WoE cũng có hạn chế:

- Dễ dẫn tới overfitting.

Với triển khai này của pandas, chúng ta cần giữ mục tiêu trong tập huấn luyện.

### Hạn chế của pandas:

- Nó không bảo toàn thông tin từ dữ liệu huấn luyện để truyền sang dữ liệu kiểm tra.

Chúng ta cần lưu trữ các mapping riêng lẻ nếu định sử dụng mô hình trong sản xuất.

```python
## Yêu cầu 1: tính tỷ lệ các hành khách sống sót
# trên tổng số người sống sót trên mỗi hạng mục của cabin

# tổng số người sống sót
## VIẾT CODE Ở ĐÂY:
total_survived = X_train['survived'].sum()

# % hành khách sống sót từ tổng số người sống sót
# mỗi hạng mục của cabin
## VIẾT CODE Ở ĐÂY:
survived = X_train.groupby(['cabin'])['survived'].sum() / total_survived

survived
```

<details><summary> Gợi ý </summary>

[sum()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sum.html)

</details>

```python
## Yêu cầu 2: tính tỷ lệ hành khách không sống sót
# trên tổng số hành khách không sống sót trên mỗi hạng mục của cabin

# tổng số hành khách không sống sót
## VIẾT CODE Ở ĐÂY:
total_non_survived = len(X_train) - X_train['survived'].sum()

# tạo một flag cho các hành khách không sống sót
## VIẾT CODE Ở ĐÂY:
X_train['non_survived'] = np.where(X_train['survived'] == 1, 0, 1)

# tính % các hành khách không sống sót
# mỗi hạng mục của cabin
## VIẾT CODE Ở ĐÂY:
non_survived = X_train.groupby(
    ['cabin'])['non_survived'].sum() / total_non_survived

non_survived
```

```python
## Yêu cầu 3: kết hợp 2 chuỗi đó thành một dataframe và tính
# WoE

# nối chuỗi trong dataframe
## VIẾT CODE Ở ĐÂY:
prob_df = pd.concat([survived, non_survived], axis=1)

# tính WoE
## VIẾT CODE Ở ĐÂY:
prob_df['woe'] = np.log(prob_df['survived']/prob_df['non_survived'])

prob_df
```

```python
## Yêu cầu 4: ghi WoE trong một dictionary

## VIẾT CODE Ở ĐÂY:
ordered_labels = prob_df['woe'].to_dict()

ordered_labels
```

```python
## Yêu cầu 5: thay các nhãn với woe

## VIẾT CODE Ở ĐÂY:
X_train['cabin'] = X_train['cabin'].map(ordered_labels)
X_test['cabin'] = X_test['cabin'].map(ordered_labels)
```

```python
## Yêu cầu 6: biến các lệnh trước đó thành 2 hàm


def find_category_mappings(df, variable, target):

    # sao chép dataframe ban đầu để không tình cờ
    # sửa nó
    tmp = df.copy()

    # tổng class dương
    ## VIẾT CODE Ở ĐÂY:
    total_pos = df[target].sum()

    # tổng class âm
    ## VIẾT CODE Ở ĐÂY:
    total_neg = len(df) - df[target].sum()

    # non-target
    ## VIẾT CODE Ở ĐÂY:
    tmp['non-target'] = 1 - tmp[target]

    # % class dương mỗi hạng mục so với tổng class dương
    pos_perc = tmp.groupby([variable])[target].sum() / total_pos

    # % class âm mỗi hạng mục so với tổng class âm
    ## VIẾT CODE Ở ĐÂY:
    neg_perc = tmp.groupby([variable])['non-target'].sum() / total_neg

    # hãy nối
    ## VIẾT CODE Ở ĐÂY:
    prob_tmp = pd.concat([pos_perc, neg_perc], axis=1)

    # tính WoE
    ## VIẾT CODE Ở ĐÂY:
    prob_tmp['woe'] = np.log(prob_tmp[target]/prob_tmp['non-target'])

    return prob_tmp['woe'].to_dict()


def integer_encode(train, test, variable, ordinal_mapping):

    ## VIẾT CODE Ở ĐÂY:
    
    train[variable] = train[variable].map(ordinal_mapping)

    test[variable] = test[variable].map(ordinal_mapping)
```

```python
## Yêu cầu 7: chạy một vòng lặp qua các biến hạng mục còn lại

for variable in ['sex', 'embarked']:
    ## VIẾT CODE Ở ĐÂY:
    mappings = find_category_mappings(X_train, variable, 'survived')
    ## VIẾT CODE Ở ĐÂY:
    integer_encode(X_train, X_test, variable, mappings)
```
