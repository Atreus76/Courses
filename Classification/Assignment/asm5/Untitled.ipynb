```python
X = df.drop(['RainTomorrow'], axis=1)
y = df['RainTomorrow']
```

```python
categorical = [var for var in X.columns if X[var].dtype=='O']
categorical
```

```python
numerical = [var for var in X.columns if X[var].dtype!='O']
numerical
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
for df1 in [X_train, X_test]:
    for col in numerical:
        col_median = X_train[col].median()
        df1[col].fillna(col_median, inplace=True)
```

```python
X_train[numerical].isnull().sum()
```

```python
for df2 in [X_train, X_test]:
    for col in cat1:
        df2[col].fillna(X_train[col].mode()[0], inplace=True)
```

```python
X_train[categorical].isnull().sum()
```

```python
X_train_cate =  X_train[categorical].drop(['Date'], axis=1)
X_train_cate.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
X_train_cate = pd.get_dummies(X_train_cate)
X_train_cate
```

```python
X_test_cate =  X_test[categorical].drop(['Date'], axis=1)
X_test_cate = pd.get_dummies(X_test_cate)
X_test_cate
```

```python
X_train_num = X_train[numerical]
X_test_num = X_test[numerical]
```

```python
train = pd.concat([X_train_num, X_train_cate], axis=1)
test = pd.concat([X_test_num, X_test_cate], axis=1)
```

```python
from sklearn.metrics import f1_score
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(train, y_train)
```

```python
y_pred = model.predict(test)
y_pred
```

```python
score = f1_score(y_test, y_pred)
```
