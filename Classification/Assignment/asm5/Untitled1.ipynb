```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('dataset/weatherAUS.csv')
```

```python
data.head()
```

```python
X = data.drop(['Date', 'RainTomorrow'], axis=1)
y = data['RainTomorrow']
```

```python
data.info()
```

```python
categorical = [var for var in X.columns if X[var].dtype == 'O']
```

```python
numerical = [var for var in X.columns if X[var].dtype != 'O']
```

```python
X[categorical].head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
for df in [X_train, X_test]:
    for col in numerical:
        col_median = df[col].median()
        df[col].fillna(col_median, inplace=True)
```

```python
X_train[numerical].isnull().sum()
```

```python
X_train_categorical = X_train[categorical]
```

```python
X_train_categorical = pd.get_dummies(X_train_categorical)
```

```python
X_train_categorical.head()
```

```python
X_test_categorical = X_test[categorical]
X_test_categorical = pd.get_dummies(X_test_categorical)
```

```python
X_train_numerical = X_train[numerical]
X_test_numerical = X_test[numerical]
```

```python
train = pd.concat(X_train_categorical, X_train_numerical)
test = pd.concat(X_test_categorical, X_test_numerical)
```
